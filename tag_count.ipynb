```python
for d in tqdm(list(db1.find())):
    for t in d["topics"]:
        if t in tag_map:
            tag_map[t] += 1
        else:
            tag_map[t] = 1
        
        tag = st.stem(t)
        if tag in stem_tag_map:
            stem_tag_map[tag] += 1
        else:
            stem_tag_map[tag] = 1
```
